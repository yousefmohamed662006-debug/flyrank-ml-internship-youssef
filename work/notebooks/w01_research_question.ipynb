## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

The starter dataset contains 30,000 pages from 32 pseudonymized clients. An observed 54.2% of the pages are labelled as declining. This suggests that the content team has a large prioritization problem: there are many pages to review, but limited time to decide which pages should be investigated first.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
from pathlib import Path
import subprocess
import sys
import pandas as pd

# Locate the starter repository and dataset.
if "google.colab" in sys.modules:
    repo_dir = Path("/content/flyrank-ml-internship-starter")

    if not repo_dir.exists():
        subprocess.run(
            [
                "git",
                "clone",
                "--depth",
                "1",
                "https://github.com/flyrank-bih/flyrank-ml-internship-starter",
                str(repo_dir),
            ],
            check=True,
        )
else:
    repo_dir = Path.cwd()

    # Handle running the notebook from work/notebooks/.
    if repo_dir.name == "notebooks":
        repo_dir = repo_dir.parent.parent

csv_path = repo_dir / "data" / "raw" / "content_refresh_anonymized.csv"

if not csv_path.exists():
    raise FileNotFoundError(f"Could not find the starter CSV at: {csv_path}")

df = pd.read_csv(csv_path)

# Three real numbers that describe the opportunity.
total_pages = len(df)
total_clients = df["client_id"].nunique()
declining_rate = (
    df["trend_direction"]
    .astype(str)
    .str.lower()
    .eq("down")
    .mean()
    * 100
)

print(f"1. Total pages: {total_pages:,}")
print(f"2. Pseudonymized clients: {total_clients:,}")
print(f"3. Observed declining rate: {declining_rate:.1f}%")

print("\nMarkdown paragraph:")
print(
    f"The starter dataset contains {total_pages:,} pages from "
    f"{total_clients:,} pseudonymized clients. "
    f"An observed {declining_rate:.1f}% of the pages are labelled as declining. "
    "This suggests that the content team has a large prioritization problem: "
    "there are many pages to review, but limited time to decide which pages "
    "should be investigated first."
)
```
